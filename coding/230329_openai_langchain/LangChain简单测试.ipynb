```python
if not os.path.exists('cut'): 
  os.makedirs('cut')
```

```python
file_list = ['2022世界杯.txt', '安倍晋三遇刺案.txt']
text_list = [text_1, text_2]
cut_list = []

for file, text in zip(file_list, text_list):
    cut_data = " ".join([w for w in list(jb.cut(text))])
    cut_list.append(cut_data)  # 看一下切完的结果
    #分词处理后的文档保存到data文件夹中的cut子文件夹中
    cut_file= f"./cut/cut_{file}"
    with open(cut_file, 'w') as f:   
        f.write(cut_data)
        f.close()
```

```python
cut_list
```

```python
#加载文档
loader = DirectoryLoader('./cut',glob='**/*.txt')
docs = loader.load()
#文档切块
text_splitter = TokenTextSplitter(chunk_size=1000, chunk_overlap=0)
doc_texts = text_splitter.split_documents(docs)
```

```python
#调用openai Embeddings
os.environ["OPENAI_API_KEY"] = key
embeddings = OpenAIEmbeddings(openai_api_key=os.environ["OPENAI_API_KEY"])
#向量化
vectordb = Chroma.from_documents(doc_texts, embeddings, persist_directory="./data/cut")
vectordb.persist()
#创建聊天机器人对象chain
chain = ChatVectorDBChain.from_llm(OpenAI(temperature=0, model_name="gpt-3.5-turbo"), vectordb, return_source_documents=True)
```

```python
def get_answer(question):
  chat_history = []
  result = chain({"question": question, "chat_history": chat_history});
  return result["answer"]
```

```python
question = "2022年的世界杯冠军是谁？"
print(get_answer(question))
```
